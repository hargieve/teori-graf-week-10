```python
# Algoritma Pewarnaan Graf Welsh–Powell

def welsh_powell(graph):
    # 1. Urutkan simpul berdasarkan derajatnya (jumlah tetangga), dari besar ke kecil
    sorted_vertices = sorted(graph, key=lambda x: len(graph[x]), reverse=True)

    # 2. Inisialisasi dictionary untuk menyimpan hasil warna
    color_assignment = {}

    # 3. Inisialisasi warna pertama
    color = 0

    # 4. Proses pewarnaan
    for vertex in sorted_vertices:
        if vertex not in color_assignment:
            color += 1
            color_assignment[vertex] = color
            # Coba beri warna yang sama ke simpul lain yang tidak bersebelahan
            for other_vertex in sorted_vertices:
                if other_vertex not in color_assignment:
                    # Jika tidak ada tetangga dengan warna yang sama
                    if not any(neigh in color_assignment and color_assignment[neigh] == color 
                               for neigh in graph[other_vertex]):
                        color_assignment[other_vertex] = color

    return color_assignment


# Contoh graf (tak berarah)
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'C', 'D']
}

# Jalankan algoritma
hasil_warna = welsh_powell(graph)

# Tampilkan hasil
print("Hasil Pewarnaan Graf (Welsh–Powell):")
for simpul, warna in hasil_warna.items():
    print(f"Simpul {simpul} ---> Warna {warna}")
```
