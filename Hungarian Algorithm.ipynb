```python
from scipy.optimize import linear_sum_assignment

# Matriks biaya (cost matrix)
cost = [
    [9, 2, 7],
    [6, 4, 3],
    [5, 8, 1]
]

# Jalankan Hungarian Algorithm
row_ind, col_ind = linear_sum_assignment(cost)

# Tampilkan hasil
print("Penugasan optimal:")
for i in range(len(row_ind)):
    print(f"Pekerja {row_ind[i]+1} → Tugas {col_ind[i]+1}")

# Total biaya minimum
total_cost = sum(cost[row_ind[i]][col_ind[i]] for i in range(len(row_ind)))
print("Total biaya minimum:", total_cost)
```
